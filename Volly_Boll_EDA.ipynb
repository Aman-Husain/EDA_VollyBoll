```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid", palette="pastel")  # Or use "deep", "dark", "muted", etc.
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['#FF5733'])  # Apply orange globally
```

```python
df = pd.read_csv('/content/VNL2023.csv')
```

```python
# Columns correlation checking
# (1 means strong +ve relationship(Both the values increasing together)
# (-1 means strong negative realtionship(If one is increasing so another one is decreasing))
numeric_cols = df.select_dtypes(include = ['int','float']).columns
corr_matrix = df[numeric_cols].corr()
print(corr_matrix)
```

```python
# Having visual representions using heatmap
plt.figure(figsize = (8,6))
sns.heatmap(corr_matrix,annot = True,cmap = 'coolwarm',linewidth = 5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Value counts for position columns
position_counts = df['Position'].value_counts()
position_counts
```

```python
# Distribution of positions
plt.pie(position_counts,labels = position_counts.index,autopct = '%1.2f%%',startangle = 90)
plt.title('Distribution Of Positions')
plt.show()
```

```python
# Average attack of each country
avg_attack_by_country = df.groupby('Country')['Attack'].mean()
avg_attack_by_country.sort_values(ascending = False)
```

```python
# Visualization of attack by countries
avg_attack_by_country.sort_values(ascending = False).head(5).plot(kind = 'bar')
plt.title('Average Attack By Country')
plt.xlabel('Country')
plt.ylabel('Average Attack')
plt.show()
```

```python
# Groupby on age and calculate the average serve on each age
avg_serve_by_age = df.groupby('Age')['Serve'].mean()
avg_serve_by_age.sort_values(ascending = False)
```

```python
# Groupby to check the leading country in terms of how good they are in attack
df.groupby(['Country','Position'])['Attack'].max().reset_index().sort_values(ascending = False,by = 'Attack')
```

```python
# Total dig of each country
df.groupby('Country')['Dig'].sum()
```

```python
# Scatter block on Block and receive
plt.scatter(df['Block'],df['Receive'])
plt.title('Scatter Plot : Block vs Received')
plt.xlabel('Block')
plt.ylabel('Receive')
plt.show()
```

```python
# Box plot for the distribution of serve values
sns.boxplot(x = df['Serve'])
plt.title('Box Plot:Distribution Of Serve Values')
plt.xlabel('Serve')
plt.show()
```

```python
# Histogram for the distribution of age with age groups
plt.hist(df['Age'],bins = 20,color = 'skyblue',edgecolor = 'black')
plt.title('Distribution Of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Creating bar chart for the avg attack by position
avg_attack_by_position = df.groupby('Position')['Attack'].mean()
avg_attack_by_position.sort_values(ascending = True).plot(kind = 'bar',color = 'green')
plt.title('Average Attack By Position')
plt.xlabel('Position')
plt.ylabel('Average Attack')
plt.show()
```

```python
# Line plot
serve_trend_by_age = df.groupby('Age')['Serve'].mean()
serve_trend_by_age.plot(kind = 'line',marker = 'o',linestyle = '-',color = 'gray')
plt.title('Serve Trend Over Age groups')
plt.xlabel('Age')
plt.ylabel('Average Serve')
plt.show()
```

```python
# Total Attack Block By Country
total_attack_block_by_country = df.groupby('Country')[['Attack','Block']].sum()
total_attack_block_by_country.sort_values(ascending = False,by = 'Attack').plot(kind =  'bar',stacked = True,colormap = 'viridis')
plt.title('Total Attack And Block By Country')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.show()
```
